```python
from imblearn.under_sampling import RandomUnderSampler

X = PASS_TEMP.drop('outcome', axis=1)
y = PASS_TEMP['outcome']

rus = RandomUnderSampler(random_state=42)

X_resampled, y_resampled = rus.fit_resample(X, y)

resampled_PASS_TEMP = pd.DataFrame(X_resampled, columns=X.columns)
resampled_PASS_TEMP['outcome'] = y_resampled

COUNT = pd.DataFrame()
COUNT['Pass frequency'] = list(resampled_PASS_TEMP.outcome.value_counts())
COUNT['Pass outcome'] = resampled_PASS_TEMP.outcome.value_counts().index

fig, ax = plt.subplots(figsize = (5,5))
fig.set_facecolor('lightgrey')
ax.set_facecolor('whitesmoke')
sns.barplot(x = 'Pass outcome', y = 'Pass frequency', data = COUNT, ax = ax, hue = 'Pass outcome', palette = 'husl')
ax.text(x=-0.296, y=20000, s = f"{int(100*COUNT.loc[0]['Pass frequency']/len(resampled_PASS_TEMP))} %",
        weight = 'bold', color = 'white', fontsize = 30)
ax.text(x=1-0.296, y=20000, s = f"{100 -int(100*COUNT.loc[0]['Pass frequency']/len(resampled_PASS_TEMP))} %",
        weight = 'bold', color = 'white', fontsize = 30)
ax.set_title("Outcome Classes \nafter UnderSampling", fontsize = 15, weight = 'bold')
ax.set_ylabel('Pass frequency', fontsize = 12, weight = 'bold')
ax.set_xlabel('Pass outcome', fontsize = 12, weight = 'bold')
ax.set_xticklabels(['complete','incomplete'], fontsize=12)
plt.show()
```

## **Correlation Heatmap**

```python
corr_table = round(shot_event[['xg', 'goal_distance', 'angle', 'teammates in triangle', 'opponents in triangle']].corr(method = 'pearson'), 2)
corr_table
```

```python
shot_event_num = shot_event[['xg', 'goal_distance', 'angle', 'teammates in triangle', 'opponents in triangle']]
corr_table = round(shot_event_num.corr(method = 'pearson'), 2)
corr = np.tril(np.ones((5,5)))*np.array(corr_table)
corr = np.where(corr != 0, corr, np.nan)
corr = pd.DataFrame(corr,
                    columns = ['xg', 'goal_distance', 'angle', 'teammates\nin triangle', 'opponents\nin triangle'],
                    index = ['xg', 'goal_distance', 'angle', 'teammates\nin triangle', 'opponents\nin triangle'])
fig, ax = plt.subplots(figsize = (12,12))
fig.patch.set_facecolor('lightyellow')
ax.set_facecolor('lightyellow')
ax.set_title("Correlation Plot\nFor Numerical Features of xG dataset", fontsize = 20, fontweight = 'semibold')
corr_plot = sns.heatmap(corr, annot = True, ax = ax, cmap='coolwarm',
                        linecolor = 'gray',linewidths = 1,
                        annot_kws = {'fontweight': 'semibold', 'fontsize': 15},
                        cbar_kws = {"orientation": "vertical",
                                    "label": "Correlation Coefficient"})

colorbar = corr_plot.collections[0].colorbar
colorbar.ax.tick_params(labelsize=12)  # Set tick label size
colorbar.set_label("Correlation Coefficient", fontsize=14, fontweight = 'semibold')  # Set label size

for label in colorbar.ax.get_yticklabels():  # For vertical colorbar
    label.set_fontweight('semibold')

# Correct the xtick label fontweight
ax.set_xticklabels(ax.get_xticklabels(), fontweight='semibold', fontsize=15)
# Correct the ytick label fontweight (if needed)
ax.set_yticklabels(ax.get_yticklabels(), fontweight='semibold', fontsize=15)
plt.tight_layout()
plt.show()
```
